### Step1 : Import **``student_marks_data.csv``** into Databricks tables folder

### Step2 :  Import SQLContext

```python
from pyspark.sql import SQLContext
```

### Step3 :  Load the  and creating DataFrame

```python
fileName="dbfs:/FileStore/tables/student_marks_data.csv"
studentMarksDF= spark.read.csv(fileName,inferSchema = "true",header="true")
studentMarksDF.show(5)
                   
```

### Step4 : Print the schema using printSchema

```python
studentMarksDF.printSchema( )
```

### Step5 : Describe the Schema

```python
display(studentMarksDF.describe())
```

### Step5 :  Adding a New Coloumn

```python
#to perform exp() operation on existing coloumn
from pyspark.sql.functions import exp

#returns a new dataframe with an additional new coloumn with name "expTest1"
df_with_exp = studentMarksDF.withColumn("expTest1",exp("test1"))
#displays first 2 records of new dataframe
df_with_exp.show(2)
```

### Step6 :  Calculating Percentage of marks of every student
This Box Plot are used to represent the Percentage score according to the Student Id.

```python
df=studentMarksDF
df_with_percentage = df.withColumn("Percentage",(df.test1+df.test2+df.test3+df.test4)/4)
display(df_with_percentage)
```

### Step7 :  Plotting Bar Graph to represent number of Students Pass or Fail
This bar graph is used to represent Number of Pass Vs Fail Students.

```python
from pyspark.sql import functions as F
k=df_with_percentage.select(df.Student_id, F.when(df_with_percentage.Percentage > 50.0 ,"Pass").otherwise("Fail").alias("Result"))
k.createOrReplaceTempView("Result1")
display(spark.sql("select Result,count(Result) from Result1 GROUP BY Result"))
```

### Step8 :  Sorting a column
This Bar Graph is used to represents how many number of students got a particular percentage.

```python
df_with_percentage.createOrReplaceTempView("Percentage_Table")
display(spark.sql("Select Student_id,Subject_type,Subject_code,year_id,Percentage from Percentage_Table where Subject_type = 'PRC' and Subject_code = 'CHE' and year_id = 2015 ORDER BY Percentage"))
```
